```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
print(df.head())
```

```python
desired_order = ['customer', 'st', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']
df = df[desired_order + list(df.columns.difference(desired_order))]
print(df)
```

```python
data = pd.DataFrame(df)
data1 = pd.DataFrame(df1)
data2 = pd.DataFrame(df2)
def get_numerical_columns(df):
    return df.select_dtypes(include='number').columns
numerical_columns_df = get_numerical_columns(df)
numerical_columns_df1 = get_numerical_columns(df1)
numerical_columns_df2 = get_numerical_columns(df2)

print("Numerical columns in df:", numerical_columns_df)
print("Numerical columns in df1:", numerical_columns_df1)
print("Numerical columns in df2:", numerical_columns_df2)
```

```python
def get_categorical_columns(df):
    return df.select_dtypes(include='object').columns

dataframes = [df, df1, df2]

for idx, df in enumerate(dataframes, start=1):
    categorical_columns = get_categorical_columns(df)
    print(f"Categorical columns in df{idx}:", categorical_columns)
```

```python
concatenated_df = pd.concat([df, df1, df2], ignore_index=True)
print(concatenated_df)
```
